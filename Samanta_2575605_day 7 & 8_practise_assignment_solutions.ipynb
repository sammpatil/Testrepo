```python
1.Write a Python program to find a target values in a list using linear search with the following steps 
a. Initialize the list to store the input elements 
b. Initialize found equal to false 
c. Enter the item to be searched bracket match underscore item 
d. For each element in the list 
1. If match underscore item equal to value
a. Return match item position 
e. If the match item is not in the list display the error message that the item is not found in the list
```

```python
#Step a: Initialize the list
input_list = [5, 10, 15, 20, 25]

# Step b: Initialize found to false
found = False

# Step c: Enter the item to be searched
match_item = int(input("Enter the item to be searched: "))

# Step d: For each element in the list
for index, item in enumerate(input_list):
    # Step d.1: If match_item is equal to the current item
    if item == match_item:
        # Step d.1a: Return match_item position
        print(f"Item found at position {index}")
        found = True
        break

# Step e: If match_item is not in the list, display an error message
if not found:
    print("Item not found in the list.")
```

```python
2.Write a Python program to implement binary search to find the target values from the list.
a. Create separate functions to do binary search. 
b. Get the number of inputs from the user. 
c. Store the inputs individually in a list. 
d. In a binary search function, at first sort the list in order to start the search from the middle of the list. 
e. Compare the middle element to the right and the left element to search the target element. 
f. If greater, move to right of a list or else move to another side of the list. 
g. Print the result along with the position of the element.
```

```python
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid  # Target found, return its position
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1  # Target not found in the list

# Step b: Get the number of inputs from the user
num_inputs = int(input("Enter the number of inputs: "))

# Step c: Store the inputs individually in a list
input_list = []
for i in range(num_inputs):
    input_value = int(input(f"Enter input {i + 1}: "))
    input_list.append(input_value)

# Sort the input list for binary search
input_list.sort()

# Prompt the user to enter the target value
target_value = int(input("Enter the target value to search for: "))

# Step e: Call binary_search function to find the target element
result = binary_search(input_list, target_value)

# Step g: Print the result along with the position of the element
if result != -1:
    print(f"Item {target_value} found at position {result}")
else:
    print(f"Item {target_value} not found in the list.")
```

```python
3.Write a Python program for sorting a list of elements using selection sort algorithm. 
a. Assume two sorted lists initially empty and unsorted list given input list. 
b. In the first iteration, find the smallest element in the unsorted list and place it in the sorted list. 
c. In the second iteration, find the smallest element in the unsorted list and place it in the correct position by
comparing with the element in the sorted list.
d. In the third iteration, again find the smallest element in the unsorted list and place it in the correct position 
by comparing with the element in the sorted list. 
e. This process continues till the unsorted list becomes empty. 
f. Display the sorted list.
```

```python
#Selection Sort Function
def selection_sort(arr):
    sorted_list = []  # Initialize the sorted list as empty
    while arr:  # Continue until the unsorted list becomes empty
        smallest = min(arr)  # Find the smallest element in the unsorted list
        sorted_list.append(smallest)  # Place it in the sorted list
        arr.remove(smallest)  # Remove the smallest element from the unsorted list
    return sorted_list

# Input list
input_list = [64, 25, 12, 22, 11]

# Step-by-step sorting
for i in range(len(input_list)):
    sorted_list = selection_sort(input_list[:i + 1])
    print(f"Iteration {i + 1}: {sorted_list}")

# Final sorted list
final_sorted_list = selection_sort(input_list)
print("Sorted list:", final_sorted_list)
```

```python
4.Write a python program for sorting a list of elements using insertion sort algorithm 
A. Assume two lists, sorted list, initially empty and unsorted list, given input list. B. In the first iteration, take 
the first element in the unsorted list and insert it in the sorted list.
C in the second iteration, take the second element in the given list and compare with the element in the sorted sublist 
and place it in the correct position.
D. In the third iteration, take the third element in the given list and compare with the elements in the sorted sub-list
and place the elements in the correct position.
E. This process continues until the last element is inserted in the sorted sublist.
F. Display the sorted elements.
```

```python
def insertion_sort(input_list):
    sorted_list = []
    for unsorted_element in input_list:
        if not sorted_list:
            sorted_list.append(unsorted_element)
        else:
            for i in range(len(sorted_list)):
                if unsorted_element < sorted_list[i]:
                    sorted_list.insert(i, unsorted_element)
                    break
            else:
                sorted_list.append(unsorted_element)
    return sorted_list

# Example usage:
input_list = [5, 2, 9, 1, 5, 6]
sorted_result = insertion_sort(input_list)
print("Sorted List:", sorted_result)
```

```python
5. Write a Python program that performs merge sort on a list of numbers:

a. Divide: If the given array has zero or one element, return.

1. Otherwise

ii. Divide the input list in to two halves each containing half of the

elements. i.e. left half and right half.

b. Conquer: Recursively sort the two lists (left half and right half).

a. Call the merge sort on left half.

b. Call the merge sort on right half.

Combine: Combine the elements back in the input list by merging the two sorted lists into a sorted sequence.
```

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    # Divide the input list into two halves
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    # Recursively sort the two halves
    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    # Combine the sorted halves back into a sorted sequence
    sorted_arr = merge(left_half, right_half)
    
    return sorted_arr

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    
    return result

# Input list of numbers
input_list = [38, 27, 43, 3, 9, 82, 10]

# Call the merge_sort function to sort the list
sorted_list = merge_sort(input_list)

# Display the sorted list
print("Sorted list:", sorted_list)
```

```python
6.Write a Python script to perform the following operation on a singly linked list.
A. Create a list. 
B. Find the smallest element from the list. 
C. Insert an element if it is not a duplicate element. 
D. Display the elements in the reverse order.
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def find_smallest(self):
        if not self.head:
            return None
        smallest = self.head.data
        current = self.head.next
        while current:
            if current.data < smallest:
                smallest = current.data
            current = current.next
        return smallest

    def insert_unique(self, data):
        if not self.head:
            self.head = Node(data)
            return
        current = self.head
        while current:
            if current.data == data:
                return  # Element is already in the list
            if current.next is None:
                current.next = Node(data)
                return
            current = current.next

    def display_reverse(self):
        if not self.head:
            return []
        values = []
        current = self.head
        while current:
            values.append(current.data)
            current = current.next
        return values[::-1]

# Example usage:
linked_list = LinkedList()
linked_list.append(5)
linked_list.append(3)
linked_list.append(8)
linked_list.append(3)
linked_list.append(2)

print("Original List:", linked_list.display_reverse())

smallest_element = linked_list.find_smallest()
print("Smallest Element:", smallest_element)

linked_list.insert_unique(7)
linked_list.insert_unique(3)  # Duplicate, won't be inserted

print("List After Insertions:", linked_list.display_reverse())
```

```python
7.Write a Python program to implement the various operations for Stack ADT. 
1.Push
2.Pop
3.Display
```

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return "Stack is empty"

    def display(self):
        if not self.is_empty():
            for item in reversed(self.items):
                print(item)
        else:
            print("Stack is empty")

    def is_empty(self):
        return len(self.items) == 0

# Create a new stack
stack = Stack()

# Push elements onto the stack
stack.push(1)
stack.push(2)
stack.push(3)

# Display the stack
print("Stack elements:")
stack.display()

# Pop an element from the stack
popped_item = stack.pop()
print("Popped item:", popped_item)

# Display the stack again
print("Stack elements after popping:")
stack.display()
```

```python
8.Write a Python script to implement the various operations for Queue ADT. 
1. Insert
2. Delete 
3. Display
```

```python
class Queue:
    def __init__(self):
        self.items = []

    def insert(self, item):
        self.items.append(item)

    def delete(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            raise Exception('Queue is empty - cannot delete')

    def is_empty(self):
        return len(self.items) == 0

    def display(self):
        if not self.is_empty():
            print("Queue Contents:")
            for item in self.items:
                print(item)
        else:
            print("Queue is empty.")

# Create a queue
queue = Queue()

# Insert elements into the queue
queue.insert(1)
queue.insert(2)
queue.insert(3)

# Display the queue
queue.display()

# Delete elements from the queue
deleted_item = queue.delete()
print(f"Deleted item: {deleted_item}")

# Display the queue after deleting
queue.display()
```

```python
9. Write a program in python to convert the following infix expression to its postfix form using push and pop 
operations of a Stack

a. A/B^C+D*E-F*G

b. (B^2-4*A*C)^(1/2) (100)
```

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            raise Exception('Stack is empty - cannot pop')

    def is_empty(self):
        return len(self.items) == 0

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return None

def infix_to_postfix(expression):
    precedence = {'+': 1, '-': 1, '*': 2, '/': 2, '^': 3}

    stack = Stack()
    postfix = []

    for char in expression:
        if char.isalnum():
            postfix.append(char)
        elif char == '(':
            stack.push(char)
        elif char == ')':
            while not stack.is_empty() and stack.peek() != '(':
                postfix.append(stack.pop())
            stack.pop()  # Pop the left parenthesis
        else:
            while (not stack.is_empty() and
                   stack.peek() != '(' and
                   precedence.get(char, 0) <= precedence.get(stack.peek(), 0)):
                postfix.append(stack.pop())
            stack.push(char)

    while not stack.is_empty():
        postfix.append(stack.pop())

    return ''.join(postfix)

# Test cases
expression_a = "A/B^C+D*E-F*G"
expression_b = "(B^2-4*A*C)^(1/2) (100)"

postfix_a = infix_to_postfix(expression_a)
postfix_b = infix_to_postfix(expression_b)

print("Postfix expression (a):", postfix_a)
print("Postfix expression (b):", postfix_b)
```
