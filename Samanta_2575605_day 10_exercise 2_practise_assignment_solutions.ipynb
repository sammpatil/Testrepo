# 3. Plot the number of films that have been released each decade over the history of cinema.

```python
import pandas as pd
import matplotlib.pyplot as plt

yearly_movie_counts = titles['year'].value_counts().reset_index()
yearly_movie_counts.columns = ['year', 'count']

# Define a function to round the year to the nearest decade
def round_to_decade(year):
    return int(year // 10 * 10)

# Apply the rounding function to create a 'decade' column
yearly_movie_counts['decade'] = yearly_movie_counts['year'].apply(round_to_decade)

# Group the data by decade and sum the counts
decade_movie_counts = yearly_movie_counts.groupby('decade')['count'].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.bar(decade_movie_counts['decade'], decade_movie_counts['count'], width=5, align='center')
plt.xlabel('Decade')
plt.ylabel('Number of Films Released')
plt.title('Number of Films Released Each Decade Over the History of Cinema')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 4. Plot the number of "Hamlet" films made each decade.

```python
import pandas as pd
import matplotlib.pyplot as plt

hamlet_movies = titles[titles['title'] == 'Hamlet']

# Create a DataFrame with the count of "Hamlet" films released each year
hamlet_yearly_counts = hamlet_movies['year'].value_counts().reset_index()
hamlet_yearly_counts.columns = ['year', 'count']

# Define a function to round the year to the nearest decade
def round_to_decade(year):
    return int(year // 10 * 10)

# Apply the rounding function to create a 'decade' column
hamlet_yearly_counts['decade'] = hamlet_yearly_counts['year'].apply(round_to_decade)

# Group the data by decade and sum the counts
hamlet_decade_counts = hamlet_yearly_counts.groupby('decade')['count'].sum().reset_index()

# Create a bar plot
plt.figure(figsize=(12, 6))
plt.bar(hamlet_decade_counts['decade'], hamlet_decade_counts['count'], width=5, align='center')
plt.xlabel('Decade')
plt.ylabel('Number of "Hamlet" Films Released')
plt.title('Number of "Hamlet" Films Released Each Decade')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 12. Plot how many roles Mammootty has played in each year of his career.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from 'cast.csv' (assuming the CSV file has relevant columns)
cast = pd.read_csv('cast.csv')

# Filter for roles played by Mammootty
mammootty_roles = cast[cast['name'] == 'Mammootty']

# Group the roles by year and count the number of roles in each year
roles_by_year = mammootty_roles.groupby('year').size().reset_index(name='Number of Roles')

# Create a line plot
plt.figure(figsize=(12, 6))
plt.plot(roles_by_year['year'], roles_by_year['Number of Roles'], marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('Number of Roles')
plt.title('Number of Roles Played by Mammootty Each Year')
plt.grid(True)
plt.show()
```
